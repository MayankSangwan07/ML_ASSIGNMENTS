```python
import numpy as np
```

```python
# (a) Reverse the NumPy array
arr = np.array([1,2,3,6,4,5])

print("Orginal Array : ",arr)
rev_arr = arr[::-1]
print("Reverse : ", rev_arr)
```

```python
# (b) Flatten using two methods
arr = np.array([[1,2,3],[2,4,5],[1,2,3]])
flat1 = arr.flatten()
flat2 = arr.ravel()
print("flaten - method 1 : ",flat1)
print("flaten - method 2 : ",flat2)
```

```python
# (c) Compare arrays
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[1,2],[3,4]])
print("Array equal? :",np.array_equal(arr1,arr2))
```

```python
# (d) Most frequent value and indices
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1,1,1,2,3,4,2,4,3,3])

def most_freq(arr):
  vals,counts = np.unique(arr,return_counts=True)
  max_count = np.max(counts)
  most_val = vals[counts == max_count]
  indices = [np.where(arr == v)[0] for v in most_val]
  return most_val, indices

print("Most frequent value in x : ",most_freq(x))
print("Most frequent value in y : ",most_freq(y))
```

```python
# (e) Matrix sums
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Total sum:",gfg.sum())
print("Row-wise sum:",np.sum(gfg,axis=1))
print("Column-wise sum:",np.sum(gfg,axis=0))
```

```python
# (f) Matrix operations
n_array = np.array([[55,25,15],[30,44,2],[11,45,77]])
print("Diagonal sum: ",np.trace(n_array))
eig_vals,eig_vecs = np.linalg.eig(n_array)
print("Eigen values: ",eig_vals)
print("Eigen vectors: ",eig_vecs)

print("Inverse: ",np.linalg.inv(n_array))
print("Determinant: ",np.linalg.det(n_array))
```

```python
# (g) Matrix multiplication & covariance
p1 = np.array([[1,2],[2,3]])
q1 = np.array([[4,5],[6,7]])
print("Mat Mul: ",np.dot(p1,q1))
print("Covariance: ",np.cov(p1,q1))

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
print("Mat Mul: ",np.dot(p2,q2))
print("Covariance: ",np.cov(p2.T,q2))
```

```python
# (h) Inner, outer, cartesian product
x = np.array([[2,3,4], [3,2,9]])
y = np.array([[1,5,0], [5,10,3]])

print("Inner Product: ",np.inner(x,y))
print("Outer Product: ",np.inner(x,y))
print("Cartesian Product: ",np.array(np.meshgrid(x,y)))
```

```python
# (a) Array operations

array = np.array([[1,-2,3],[-4,5,-6]])
print("Absolute value: ",np.abs(array))

flat = array.flatten()
print("percentiles(flat): ",np.percentile(flat,[25,50,75]))
print("precentiles(col): ",np.percentile(array,[25,50,75],axis=0))
print("precentiles(row): ",np.percentile(array,[25,50,75],axis=1))


print("Mean(flat): ",np.mean(flat))
print("Median(flat): ",np.median(flat))
print("Std Dev(flat): ",np.std(flat))
print("Mean(col): ",np.mean(flat,axis=0))
print("Median(col): ",np.median(flat,axis=0))
print("Std Dev(col): ",np.std(flat,axis=0))
print("Mean(row): ",np.mean(array,axis=1))
print("Median(row): ",np.median(array,axis=1))
print("Std Dev(row): ",np.std(array,axis=1))
```

```python
# (b) Floor, ceil, trunc, round
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Floor: ",np.floor(a))
print("Ceil: ",np.ceil(a))
print("Trunc: ",np.trunc(a))
print("Round: ",np.round(a))
```
